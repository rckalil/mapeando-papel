```python
import pandas as pd
from googletrans import Translator
import time

# Ler o arquivo CSV
data = pd.read_csv('../flow_heat/big.csv')

# Instanciar o tradutor
tradutor = Translator()

# Função para traduzir os nomes dos países
def traduzir_pais(pais):
    tentativas = 3
    for tentativa in range(tentativas):
        try:
            print(f'Tentativa {tentativa + 1} de traduzir: {pais}')
            traducao = tradutor.translate(pais, dest='pt').text
            print(f'Tradução: {traducao}')
            return traducao
        except Exception as e:
            print(f'Erro ao traduzir {pais}: {e}')
            time.sleep(1)  # Esperar 1 segundo antes de tentar novamente
    print(f'Não foi possível traduzir {pais} após {tentativas} tentativas.')
    return pais

# Aplicar a função de tradução à coluna 'Country'
#data['Country'] = data['Country'].apply(traduzir_pais)

# Exibir as primeiras linhas do DataFrame
print(data.head())
```

```python
import requests
import pycountry
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")

div = soup.find("div", {"class": "plainlist"})

# Find the table containing the data
table = div.find("ul")
#print(table)
# Extract the data from the table
couples = []
rows = table.find_all("li")
for row in rows:
    #print(row.text)
    cells = row.text.split("\xa0")
    #print(cells)
    country = cells[2].strip()
    alpha3_code = cells[0].strip()
    country = pycountry.countries.get(alpha_3=alpha3_code)
    country_name = country.name

    #print(country_name)
    couples.append((country_name, alpha3_code))
"""for row in rows:
    cells = row.find_all("td")
    if len(cells) >= 3:
        country = cells[0].text.strip()
        alpha3_code = cells[2].text.strip()
        couples.append((country, alpha3_code))"""

# Print the extracted data
for country, alpha3_code in couples:
    print(country, alpha3_code)
```

```python
for country, alpha3_code in couples:
    data.loc[data['Country'] == country, 'Code'] = alpha3_code
    #print(country, alpha3_code)

#print(data.head())
#print(data[data['Code'].isnull()]['Country'])
```
