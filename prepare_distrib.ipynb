```python
import pandas as pd

# Carregar seus arquivos (ex: exportacoes.csv e importacoes.csv)
df_exp = pd.read_csv('./derived/exp_1.csv')
df_imp = pd.read_csv('./derived/imp_1.csv')

def processar_fluxo(df, tipo):
    # Agrupa por País, Ano e ID
    df_agrupado = df.groupby(['Country', 'Year', 'Code']).agg({'Value': 'sum'}).reset_index()
    # Renomeia para algo como "Total Exports" ou "Total Imports"
    df_agrupado.rename(columns={'Value': tipo}, inplace=True)
    return df_agrupado

# Processar e salvar arquivos anuais (ex: country_data_until_1960.csv)
# Dica: O seu JS espera um arquivo para cada ano.
for ano in range(2016, 2025):
    df_ano_exp = processar_fluxo(df_exp[df_exp['Year'] == ano], 'Exports')
    df_ano_imp = processar_fluxo(df_imp[df_imp['Year'] == ano], 'Imports')
    
    # Merge nos dois
    df_final = pd.merge(df_ano_exp, df_ano_imp, on=['Country', 'Code'], how='outer').fillna(0)
    df_final.rename(columns={'Code': 'id'}, inplace=True)
    
    df_final.to_csv(f'./data_dist/country_data_until_{ano}.csv', index=False)
```
