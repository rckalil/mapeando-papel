```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
def sep( n: int = 80, sep: str = "="):
    print(sep*n)

def new_section(string: str, n: int = 80, sep: str = "="):
    m = len(string)
    if n < m:
        n = m + 6
    padding = (n-m+1)//2
    print(sep*padding + " " + string + " " + sep*padding)
```

```python
coffee_distribution = pd.read_csv("data/psd_coffee.csv")
coffee_distribution = coffee_distribution.dropna()
coffee_distribution.to_csv("data/psd_coffee_clean.csv", index=False)
```

```python
new_section("Coffee Distribution Head")
print(coffee_distribution.head())
sep()
```

```python
new_section("Coffee Distribution Columns")
print(coffee_distribution.columns)
sep()
```

```python
new_section("Coffee Distribution Countries")
countries = coffee_distribution["Country"].unique()
print("Número total de países:", len(countries))
print(countries)
sep()
```

```python
new_section("Coffee Distribution html list")
with open('txt_output/output_li_countries_dist.txt', 'w') as file:
    for country in countries:
        # Escrevendo cada país como um item de lista HTML no arquivo
        file.write(f'<li class="country">{country}</li>\n')
with open('txt_output/output_li_countries_dist.txt', 'r') as file:
    print(file.read())
sep()
```

```python
new_section("Coffee Distribution id")
print(len(coffee_distribution["id"].unique()))
print(coffee_distribution["id"].unique())
sep()
```

```python
new_section("Coffee Distribution Columns of Interest")
columns_interest = ["Bean Exports", "Bean Imports", "Exports","Imports", "Roast & Ground Exports","Roast & Ground Imports", "Soluble Exports","Soluble Imports", "Total Distribution","Production"]
print("Número total de colunas de interesse:", len(columns_interest))
for col in columns_interest:
    print(f"'{col}'")
sep()
```

```python
coffee_distribution_variations = coffee_distribution[columns_interest+["Country"]+["id"]+["Year"]]
```

```python
new_section("Coffee Distribution Variations Head")
print(coffee_distribution_variations.columns)
print(coffee_distribution_variations.head())
coffee_distribution_variations.to_csv("data/psd_coffee_variations.csv", index=False)
sep()
```

```python
coffee_distribution_variations.drop(columns=["Country","id"]).groupby(["Year"]).sum().to_csv("data/psd_coffee_variations_year.csv")
coffee_distribution_variations_year = pd.read_csv("data/psd_coffee_variations_year.csv")
```

```python
new_section("Coffee Distribution Variations Year Head")
print(coffee_distribution_variations_year.shape)
print(coffee_distribution_variations_year["Year"].unique())
print(coffee_distribution_variations_year.head())
sep()
```

```python
new_section("Coffee Distribution Variations Year")
plt.figure(figsize=(12, 6))
for col in columns_interest:
    sns.lineplot(x="Year", y=col, data=coffee_distribution_variations_year)
plt.legend(columns_interest)
plt.title("Coffee Distribution Variations Year")
plt.show()
```

```python
coffee_distribution_variations.drop(columns=['Year']).groupby(["Country","id"]).mean().to_csv("data/psd_coffee_variations_mean.csv")
coffee_distribution_mean = pd.read_csv("data/psd_coffee_variations_mean.csv")
```

```python
new_section("Coffee Distribution Variations Grouped Head")
print(coffee_distribution_mean.head())
print(coffee_distribution_mean.shape)
sep()
```
